```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos conocidos
cp = 3.0
focal_length = 5.0
c_prime_p_prime = 4.0
focal_length_prime = 5.0
d = 10.0
```

```python
# Coordenadas proyectadas en las imágenes
px = 2.0
px_prime = 2.5
```

```python
# Cálculos
cx = np.sqrt(cp**2 + px**2)
cx_prime = np.sqrt(c_prime_p_prime**2 + px_prime**2)

theta_x = np.arccos(px / cx)
theta_x_prime = np.arccos(px_prime / cx_prime)
```

```python
# Aplicando la Ley de los Senos
phi = np.pi - (theta_x + theta_x_prime)
ox = (d * np.sin(theta_x)) / np.sin(phi)

print("La distancia OX es:", ox)
```

```python
# Graficar los resultados
fig, ax = plt.subplots()

# Puntos en el espacio 2D
O = (0, 0)
O_prime = (d, 0)
X = (ox * np.cos(theta_x), ox * np.sin(theta_x))

# Puntos proyectados en las imágenes
x = (cp * np.cos(theta_x), cp * np.sin(theta_x))
x_prime = (c_prime_p_prime * np.cos(theta_x_prime), c_prime_p_prime * np.sin(theta_x_prime))

# Añadir los puntos al gráfico
ax.plot(*O, 'bo', label='O (Camera 1)')
ax.plot(*O_prime, 'ro', label="O' (Camera 2)")
ax.plot(*X, 'go', label='X (Object)')
ax.plot(*x, 'mo', label='x (Image 1)')
ax.plot(*x_prime, 'co', label="x' (Image 2)")

# Añadir líneas al gráfico
ax.plot([O[0], X[0]], [O[1], X[1]], 'b-', label='OX (Ray 1)')
ax.plot([O_prime[0], X[0]], [O_prime[1], X[1]], 'r-', label="O'X (Ray 2)")
ax.plot([O[0], x[0]], [O[1], x[1]], 'm--', label='Projection 1')
ax.plot([O_prime[0], x_prime[0]], [O_prime[1], x_prime[1]], 'c--', label='Projection 2')

# Etiquetas y título
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Geometría Epipolar')

# Mostrar la leyenda
ax.legend()

# Mostrar la gráfica
plt.grid()
plt.show()
```
